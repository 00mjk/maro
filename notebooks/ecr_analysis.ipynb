```python
import pandas as pd
import os
import matplotlib.pyplot as plt

log_root = '/maro_dev/examples/ecr/q_learning/single_host_mode/log/20191206'
exp_dir_list = os.listdir(log_root)
exp_performance_file_list = [(exp_dir, f'{log_root}/{exp_dir}/runner.performance.csv') for exp_dir in exp_dir_list]
print(exp_performance_file_list)
```

```python
exp_performance_ds_list = [{'tag': item[0], 'x': range(len(pd.read_csv(item[1])['total_shortage '])), 'y': pd.read_csv(item[1])['total_shortage ']} for item in exp_performance_file_list]
```

```python
print(exp_performance_ds_list)
```

```python
tags = []
fig=plt.figure(figsize=(30, 10), dpi= 80, facecolor='w', edgecolor='k')
for item in exp_performance_ds_list:
    plt.plot(item['x'], item['y'])
    tags.append(item['tag'])

plt.legend(tags)
plt.show()
```
